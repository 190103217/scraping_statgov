```python
dfs_tmp=[]
dfs_tmp_1=[]
dfs_tmp_2=[]
dfs_tmp_3=[]
```

```python
table_1=dfs_0[0::4]
for t1 in table_1:
    time.sleep(2)
    df_1=pd.DataFrame(t1)
    outside_col, inside_col = df_1.iloc[0], df_1.iloc[1]
    df_1.columns = pd.MultiIndex.from_tuples(list(zip(outside_col,inside_col)))
    df_1=df_1.drop(df_1.index[[0,1]]).reset_index(drop=True).drop('с начала месяца', axis=1, level=1).droplevel(level=0, axis=1).rename( columns={'':'region', 'за неделю': 'Социально-значимые продовольственные товары'})
    df_1=df_1.rename(columns={df_1.columns[8]: 'date'})
    df_1=df_1.melt(id_vars=['region', 'date'], var_name="type_of_goods", value_name="price_in_per")
    #df_1 = pd.concat()
    dfs_tmp.append(df_1)

dfs_tmp = pd.concat(dfs_tmp).reset_index().drop(columns=['index'])
time.sleep(2)
table_2=dfs_0[1::4]
for t2 in table_2:
    time.sleep(2)
    df_2=pd.DataFrame(t2)
    df_2= df_2.rename(columns=df_2.iloc[1]).drop(df_2.index[[0,1]]).reset_index(drop=True).rename( columns={'':'region'})
    df_2=df_2.rename(columns={df_2.columns[10]: 'date'})
    df_2=df_2.melt(id_vars=['region', 'date'], var_name="type_of_goods", value_name="price_in_per")
    #df_1 = pd.concat()
    dfs_tmp_1.append(df_2)

dfs_tmp_1 = pd.concat(dfs_tmp_1).reset_index().drop(columns=['index'])
time.sleep(2)
table_3=dfs_0[2::4]
for t3 in table_3:
    time.sleep(2)
    df_3=pd.DataFrame(t3)
    df_3= df_3.rename(columns=df_3.iloc[1]).drop(df_3.index[[0,1]]).reset_index(drop=True).rename( columns={'':'region'})
    df_3=df_3.rename(columns={df_3.columns[11]: 'date'})
    df_3=df_3.melt(id_vars=['region', 'date'], var_name="type_of_goods", value_name="price_in_per")
    #df_1 = pd.concat()
    dfs_tmp_2.append(df_3)

dfs_tmp_2 = pd.concat(dfs_tmp_2).reset_index().drop(columns=['index'])
time.sleep(2)
table_4=dfs_0[3::4]
for t4 in table_4:
    time.sleep(3)
    df_4=pd.DataFrame(t4)
    df_4= df_4.rename(columns=df_4.iloc[1]).drop(df_4.index[[0,1]]).reset_index(drop=True).rename( columns={'':'region'})
    df_4=df_4.rename(columns={df_4.columns[10]: 'date'})
    df_4=df_4.melt(id_vars=['region', 'date'], var_name="type_of_goods", value_name="price_in_per")
    #df_1 = pd.concat()
    dfs_tmp_3.append(df_4)

dfs_tmp_3 = pd.concat(dfs_tmp_3).reset_index().drop(columns=['index'])
sleep(2)

```
